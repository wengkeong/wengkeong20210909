```python
df = pd.read_csv("data.csv")
df
```

```python
df.describe()
```

```python
cars_grouped_make = df.groupby('Make')

for make, group in cars_grouped_make:
    print(make)
    print(group[['Make', 'Model', 'Year', 'MSRP']])
```

```python
# find the most expensive, cheapest and mean price for each make
cars_grouped_make = df[['Vehicle Style', 'Market Category', 'MSRP']].groupby(
                   [df['Make'], df['Model'], df['Year']])

#list(cars_grouped_make)

for make, group in cars_grouped_make:
    print(make)
    print(f'Most Expensive : ${group["MSRP"].max()}')
    print(f'Cheapest       : ${group["MSRP"].min()}')
    print(f'Mean           : ${group["MSRP"].mean()}')
    print(" ")
```

```python
# find the most expensive car
df.sort_values(by = 'MSRP', ascending = False).iloc[0]
```

```python
# returns the index of the maximum value of a given Series / column
df['MSRP'].idxmax()
```

```python
df.loc[df['MSRP'].idxmax()]
```

```python
# it is more computationally efficient to get the index first then do an .loc
# sorting a large DataFrame, e.g., 1M rows+ takes time
```

```python
# most expensive car from each maker
cars_grouped_make = df.groupby('Make')

for make, group in cars_grouped_make:
    print(make)
    print("=" * len(make)) # creates a nice underline
    most_exp_each_make = group['MSRP'].idxmax() # gives you the index of most expensive car
    print(df.iloc[most_exp_each_make])
    print("")
```

```python
# most expensive make, model and vehicle type for each make
df_result = df[['MSRP']].groupby(
            [df['Make'], df['Model'], df['Vehicle Style']]).max()

df_result
```
